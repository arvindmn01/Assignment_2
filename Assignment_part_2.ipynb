```python
df_a=pd.read_csv(r'C:\Users\Arvind\Favorites\Downloads\DS-assignment\DS - Assignment Part 2 data set\amz_com-ecommerce_sample.csv')
df_f=pd.read_csv(r'C:\Users\Arvind\Favorites\Downloads\DS-assignment\DS - Assignment Part 2 data set\flipkart_com-ecommerce_sample.csv')
```

```python
df_f=df_f[['product_name','retail_price','discounted_price']]
df_a=df_a[['product_name','retail_price','discounted_price']]
```

```python
df_f
```

```python
product_name_f=[]
retail_price_f=[]
discounted_price_f=[]
product_name_a=[]
retail_price_a=[]
discounted_price_a=[]
```

```python
for i in range(df_a.shape[0]):
    if df_f.product_name.values[i].lower()==df_a.product_name.values[i].lower():
        product_name_f.append(df_f.product_name.values[i])
        retail_price_f.append(df_f.retail_price.values[i])
        discounted_price_f.append(df_f.discounted_price.values[i])
        product_name_a.append(df_a.product_name.values[i])
        retail_price_a.append(df_a.retail_price.values[i])
        discounted_price_a.append(df_a.discounted_price.values[i])
        
```

```python
final_df=pd.DataFrame({'Product name in Flipkart':product_name_f,'Retail price in Flipkart':retail_price_f,'Discounted price in Flipkart':discounted_price_f,'Product name in Amazon':product_name_a,'Retail Price in Amazon':retail_price_a,'Discounted Price in Amazon':discounted_price_a})

```
